1.What is a Support Vector Machine (SVM)?

->SVM finds the best decision boundary (hyperplane) that separates data points of different classes with the maximum margin (i.e., the largest distance between the boundary and the nearest data points of each class).

Key terms:

Support Vectors: The data points closest to the decision boundary.

Margin: The distance between the boundary and the support vectors.

Hyperplane: The line (in 2D) or plane (in higher dimensions) that separates classes.

Types:

Linear SVM: Used when data is linearly separable.

Non-linear SVM: Uses kernel functions (like polynomial or RBF) to separate data that is not linearly separable.

2.What is the difference between Hard Margin and Soft Margin SVM?

->Hard Margin SVM:

Tries to separate classes perfectly with no misclassification.

Works only if data is linearly separable.

Can overfit if thereâ€™s even a small noise in data.

Allows some misclassifications to handle noisy or overlapping data.

Works well for non-linearly separable or noisy data.

Reduces overfitting by allowing small violations of the margin.

5.What are Support Vectors in SVM?

->Support vectors are the data points that lie closest to the decision boundary (hyperplane) in an SVM model.

They are the most critical data points because they define the position and orientation of the hyperplane.

If we remove a support vector, the decision boundary will change.

Only these points have non-zero Lagrange multipliers (Î±áµ¢ > 0).

6. What is a Support Vector Classifier (SVC)?

->It finds the best boundary (hyperplane) that separates different classes of data points with the maximum margin (the largest possible distance between the boundary and the nearest data points of each class).

some key points are:

Support Vectors: The data points that are closest to the decision boundary. They directly affect the position of the hyperplane.

Margin: The gap between the boundary and the support vectors.

Goal: Maximize this margin to make the classifier more robust.

Soft Margin SVC: Allows some misclassifications for better generalization when data isnâ€™t perfectly separable.

Kernel Trick: Used by SVC to handle non-linear data by mapping it into a higher-dimensional space.

7. What is a Support Vector Regressor (SVR)?

->A Support Vector Regressor (SVR) is the regression version of Support Vector Machine (SVM). Instead of finding a boundary to separate classes, SVR tries to fit the best line (or hyperplane) that predicts continuous values within a certain tolerance (Îµ margin).

It works by:

Ignoring small errors within the Îµ margin (called the epsilon-insensitive zone).

Minimizing errors that fall outside this margin using support vectors.

Using a kernel function (like linear, polynomial, or RBF) to handle non-linear data.

8.What is the Kernel Trick in SVM?

->The Kernel Trick in SVM is a technique used to handle non-linear data by transforming it into a higher-dimensional space where it becomes linearly separable.

Instead of explicitly transforming the data (which can be costly), the kernel trick uses a kernel function to compute the dot product in that higher space directly.

Common kernel functions:

Linear Kernel â€“ for linearly separable data

Polynomial Kernel â€“ for curved boundaries

RBF (Gaussian) Kernel â€“ for complex, non-linear patterns


9.Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?

->Linear Kernel:

It uses a straight-line decision boundary to separate the data.

Suitable when the data is linearly separable or almost linear.

Itâ€™s simple, fast to compute, and requires fewer parameters.

Works well for large feature spaces (like text classification).

However, it cannot handle non-linear relationships in the data.

Polynomial Kernel:

It creates curved decision boundaries by using polynomial functions of the input features.

Useful when the relationship between features is moderately non-linear.

Can capture feature interactions that a linear kernel misses.

The degree of the polynomial controls the flexibility of the model.

But, higher-degree polynomials may cause overfitting and increase computational cost.

RBF (Radial Basis Function) Kernel:

It maps data into an infinite-dimensional space using a Gaussian function.

Ideal for highly non-linear and complex data patterns.

Very flexible and can adapt to almost any data shape.

Requires tuning of the gamma (Î³) parameter for best performance.

More computationally expensive and may overfit if parameters are not tuned properly.


10.What is the effect of the C parameter in SVM?

->The C parameter in SVM controls the trade-off between margin width and classification error.

Hereâ€™s how it affects the model:

If C is large:

The model tries to classify all training examples correctly, even if that means having a smaller margin.

It penalizes misclassifications heavily, so the SVM becomes less tolerant to errors.

This can lead to overfitting.

If C is small:

The model allows some misclassifications to have a wider margin.

It focuses more on having a smooth and general decision boundary.

This can help avoid overfitting but may underfit if C is too low.

11.What is the role of the Gamma parameter in RBF Kernel SVM?

->The Gamma (Î³) parameter in an RBF Kernel SVM controls how far the influence of a single training example reaches.

Hereâ€™s the effect:

If Gamma is large:

Each data point has a small area of influence.

The model fits very closely around the training data.

It may capture noise and cause overfitting.

If Gamma is small:

Each data point has a large area of influence.

The decision boundary becomes smoother.

The model may underfit because it doesnâ€™t capture complex patterns.

12.What is the NaÃ¯ve Bayes classifier, and why is it called "NaÃ¯ve"?

->The NaÃ¯ve Bayes classifier is a probabilistic machine learning algorithm based on Bayesâ€™ Theorem, used mainly for classification tasks.

It predicts the class of a data point by calculating the probability of each class given the input features and choosing the class with the highest probability.

It assumes that all features are independent of each other given the class label â€” this is the â€œnaÃ¯veâ€ assumption.

Why itâ€™s called â€œNaÃ¯veâ€:
Because in reality, features are often not completely independent, but the algorithm simplifies the computation by assuming independence, which still works surprisingly well in many practical problems like spam detection, text classification, etc.

13.What is Bayesâ€™ Theorem?

->Bayesâ€™ Theorem is a fundamental rule in probability that describes how to update the probability of a hypothesis based on new evidence.

It is written as:
P(Aâˆ£B) = P(Bâˆ£A)â‹…P(A)â€‹/P(B)

Where:

P(A|B) â†’ Probability of event A given that B has occurred (posterior probability)

P(B|A) â†’ Probability of event B given A (likelihood)

P(A) â†’ Probability of event A (prior probability)

P(B) â†’ Probability of event B (evidence or normalization factor)

14.Explain the differences between Gaussian NaÃ¯ve Bayes, Multinomial NaÃ¯ve Bayes, and Bernoulli NaÃ¯ve Bayes?

->Gaussian NaÃ¯ve Bayes:

Used when the features are continuous (like height, weight, temperature).

Assumes that the data for each class follows a normal (Gaussian) distribution.

Commonly used in problems like iris classification or sensor data.

Multinomial NaÃ¯ve Bayes:

Used when the features are counts or frequencies (like word counts in text).

Often applied in text classification and spam detection.

Works best when data represents discrete counts, not continuous values.

Bernoulli NaÃ¯ve Bayes:

Used when the features are binary (0 or 1).

Ideal for data indicating presence or absence of a feature (ex:- word appears or not).

Also used in text classification, especially for binary word occurrence models.

15. When should you use Gaussian NaÃ¯ve Bayes over other variants?

->We should use Gaussian NaÃ¯ve Bayes when your features are continuous and follow a normal (Gaussian) distribution.

Itâ€™s best for:

Numeric data like height, weight, or temperature.

Sensor or real-valued features in datasets.

Cases where data is not count-based or binary.

16. What are the key assumptions made by NaÃ¯ve Bayes?

->Feature Independence:
All features are assumed to be independent of each other given the class label.

Equal Importance:
Each feature contributes equally and independently to the outcome.

Class Conditional Independence:
The probability of each feature depends only on the class, not on other features.

17.What are the advantages and disadvantages of Naive Bayes?

->Advantages of Naive Bayes:

Simple and fast to train and predict.

Works well with small datasets and high-dimensional data.

Performs well for text classification (like spam detection).

Requires less training data compared to other models.

Disadvantages of Naive Bayes:

Assumes feature independence, which is rarely true in real data.

Can give poor performance if features are highly correlated.

Struggles with zero-frequency problems (when a feature value never appears in training).

Not suitable for complex relationships between features.


18.Why is NaÃ¯ve Bayes a good choice for text classification?

->High dimensionality: It handles large vocabularies efficiently.

Fast training and prediction: Works well even on huge text datasets.

Feature independence assumption fits text data: Words in a document are mostly treated as independent features.

Performs well with small training data: Needs fewer samples to learn probabilities.

19. Compare SVM and NaÃ¯ve Bayes for classification tasks.

->Support Vector Machine (SVM):

Works by finding the best boundary (hyperplane) that separates different classes.

Performs well on both linear and non-linear data (using kernels).

Focuses on maximizing the margin between classes.

Requires feature scaling and parameter tuning (like C and Î³).

Can be computationally expensive for large datasets.

Works well when clear class boundaries exist.

Naive Bayes:

Based on Bayesâ€™ theorem and assumes independent features.

Works well for text classification and categorical data.

Fast and simple to implement.

Performs well with small datasets.

Can struggle when features are correlated.

Doesnâ€™t require heavy parameter tuning or scaling.

20. How does Laplace Smoothing help in NaÃ¯ve Bayes?

->Laplace Smoothing (also called add-one smoothing) helps in Naive Bayes by handling the zero-frequency problem.

This problem occurs when a word or feature appears in the test data but was not seen in the training data â€” which makes its probability zero, and the entire prediction fails.

Laplace smoothing fixes this by adding 1 to all feature counts, ensuring that no probability becomes zero.

```python

'''21 Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy:'''

from sklearn.datasets import load_iris
iris =load_iris()
X = iris.data
y = iris.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of SVM Classifier on Iris dataset:", accuracy)
```

```python
'''24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,n_informative=2, random_state=42, n_clusters_per_class=1)

model = SVC(kernel='poly', degree=3)
model.fit(X, y)

import numpy as np
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.6)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.title("SVM Classifier with Polynomial Kernel")
plt.show()
```

```python
'''26. Write a Python program to train a Multinomial NaÃ¯ve Bayes classifier for text classification using the 20
Newsgroups dataset.'''

from sklearn.datasets import fetch_20newsgroups
data = fetch_20newsgroups(subset='train', categories=['sci.space', 'rec.sport.hockey', 'comp.graphics'], remove=('headers', 'footers', 'quotes'))
X_train = data.data
y_train = data.target

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_vec, y_train)

test_data = fetch_20newsgroups(subset='test', categories=['sci.space', 'rec.sport.hockey', 'comp.graphics'], remove=('headers', 'footers', 'quotes'))
X_test_vec = vectorizer.transform(test_data.data)
y_pred = model.predict(X_test_vec)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_data.target, y_pred)
print("Accuracy of Multinomial Naive Bayes on 20 Newsgroups dataset:", accuracy)
```

```python
'''27.Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually'''

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                           n_informative=2, random_state=42, n_clusters_per_class=1)

C_values = [0.1, 1, 10]
plt.figure(figsize=(12, 4))

for i, C in enumerate(C_values, 1):
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(1, 3, i)
    plt.contourf(xx, yy, Z, alpha=0.6)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title(f"SVM with C = {C}")

plt.tight_layout()
plt.show()
```

```python
'''30.Write a Python program to train a Gaussian NaÃ¯ve Bayes model and compare the predictions before and
after Laplace Smoothing'''

from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.naive_bayes import GaussianNB
model1 = GaussianNB(var_smoothing=1e-9)
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_test)

model2 = GaussianNB(var_smoothing=1)
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)

from sklearn.metrics import accuracy_score
acc1 = accuracy_score(y_test, y_pred1)
acc2 = accuracy_score(y_test, y_pred2)

print("Accuracy without Laplace Smoothing:", acc1)
print("Accuracy with Laplace Smoothing:", acc2)
```

```python
'''31 Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,
gamma, kernel)'''

from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['linear', 'rbf', 'poly']
}

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=0)
grid.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
y_pred = grid.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Score:", grid.best_score_)
print("Test Accuracy after Tuning:", acc)
```

```python
'''32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and
check it improve accuracy'''

from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9, 0.1], random_state=42)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.svm import SVC
model_normal = SVC(kernel='rbf')
model_normal.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
pred_normal = model_normal.predict(X_test)
acc_normal = accuracy_score(y_test, pred_normal)

model_weighted = SVC(kernel='rbf', class_weight='balanced')
model_weighted.fit(X_train, y_train)
pred_weighted = model_weighted.predict(X_test)
acc_weighted = accuracy_score(y_test, pred_weighted)

print("Accuracy without class weighting:", acc_normal)
print("Accuracy with class weighting:", acc_weighted)
```

```python
'''35.Write a Python program to perform feature selection before training a NaÃ¯ve Bayes classifier and compare
results'''

from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.naive_bayes import GaussianNB
model1 = GaussianNB()
model1.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
pred1 = model1.predict(X_test)
acc1 = accuracy_score(y_test, pred1)

from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)

X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y, test_size=0.3, random_state=42)
model2 = GaussianNB()
model2.fit(X_train_new, y_train_new)
pred2 = model2.predict(X_test_new)
acc2 = accuracy_score(y_test_new, pred2)

print("Accuracy without Feature Selection:", acc1)
print("Accuracy with Feature Selection:", acc2)
```

```python
'''38.Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the
average accuracy.'''

from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target

from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []

for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model = SVC(kernel='rbf')
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    acc = accuracy_score(y_test, pred)
    accuracies.append(acc)

avg_acc = sum(accuracies) / len(accuracies)
print("Average Accuracy using Stratified K-Fold Cross-Validation:", avg_acc)
```

```python
'''40.Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy'''
from sklearn.datasets import load_wine
data = load_wine()
X = data.data
y = data.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
pred1 = model.predict(X_test)
acc1 = accuracy_score(y_test, pred1)

from sklearn.feature_selection import RFE
rfe = RFE(estimator=SVC(kernel='linear'), n_features_to_select=5)
rfe.fit(X_train, y_train)
X_train_rfe = rfe.transform(X_train)
X_test_rfe = rfe.transform(X_test)

model_rfe = SVC(kernel='linear')
model_rfe.fit(X_train_rfe, y_train)
pred2 = model_rfe.predict(X_test_rfe)
acc2 = accuracy_score(y_test, pred2)

print("Accuracy without RFE:", acc1)
print("Accuracy with RFE:", acc2)
```
