#Theoritical questions

1. What is unsupervised learning in the context of machine learning?

->Unsupervised learning is a type of machine learning where the model learns patterns from unlabeled data (data without predefined output).
The goal is to discover hidden structure, such as grouping, similarities, or patterns.

Key points:

No target/output variable.

Finds patterns, groups, or structure on its own.

Common tasks: clustering, dimensionality reduction, anomaly detection.

Examples:
K-Means clustering, PCA, Association Rule Mining.

2.How does K-Means clustering algorithm work?

->K-Means is an unsupervised clustering algorithm used to divide data into K clusters, where each data point belongs to the cluster with the nearest mean (centroid).

Steps of K-Means:

Choose K:
Decide how many clusters we want (K).

Initialize Centroids:
Randomly select K points as the initial cluster centers (centroids).

Assign Points to Clusters:
For each data point, calclate the distance to each centroid and assign it to the nearest one.

Update Centroids:
Recalculate each centroid as the mean of all points assigned to that cluster.

Repeat:
Keep repeating steps 3 and 4 until:

Centroids stop changing (convergence), or

Maximum iterations reached.

3.Explain the Concept of a Dendrogram in Hierarchical Clustering?

->A dendrogram is a tree-like diagram that shows how clusters are formed in hierarchical clustering.
It visualizes the process of merging (agglomerative) or splitting (divisive) clusters.

Key points:

Each leaf = a data point.

Branches show how clusters combine at different distances.

The height at which two clusters merge shows similarity/distance.

Cutting the dendrogram at a certain height gives the final clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?

->K-Means	:
Requires choosing K in advance.
Works bottom-up or top-down.
Good for large datasets	.
Produces spherical clusters.

Hierarchical Clustering :
No need to pre-define number of clusters.
K-Means works with centroid iteration.
Computationally expensive for large datasets.
Can produce non-spherical clusters.

5.What are the advantages of DBSCAN over K-Means?

->Can find arbitrary shaped clusters (not only spherical).

Does not require K to be specified in advance.

Can detect noise/outliers clearly.

Works well with clusters of different densities (K-Means fails here).

More robust to distance-based errors than K-Means.

6.When would you use Silhouette Score in clustering?

->Use the Silhouette Score to:

Measure clustering quality.

Compare different values of K (or parameters in DBSCAN).

Check how well each point fits in its assigned cluster.

Detect over-clustering or under-clustering.

Range: −1 to +1

+1 → perfect clustering

0 → overlapping clusters

−1 → wrong clustering

7.What are the limitations of Hierarchical Clustering?

->High computational cost → not suitable for large datasets.

Once a merge/split happens, cannot be undone.

Sensitive to noise and outliers.

Choice of linkage (single, complete, average) affects results.

Difficult to scale beyond thousands of points.

8. Why is feature scaling important in clustering algorithms like K-Means?

->Because K-Means uses Euclidean distance, features with larger numeric ranges dominate the distance.

Scaling ensures:

All features contribute equally.

The algorithm forms correct, balanced clusters.

Faster and more stable convergence.

Methods: Standardization, Min-Max scaling.

9. How does DBSCAN identify noise points?

->DBSCAN uses two parameters:

ε (epsilon): radius

minPts: minimum points to form a dense region

A point is a noise point if:

It has fewer than minPts neighbors within distance ε

And it is not reachable from any core point.

Types of points:

Core point: ≥ minPts neighbors

Border point: < minPts but reachable from a core

Noise point: neither core nor border

Noise points are simply treated as outliers.

10. Define inertia in the context of K-Means?

->Inertia is the sum of squared distances between each data point and its assigned cluster centroid.
It measures how compact the clusters are.

Lower inertia → tighter, better clusters

Higher inertia → poorer clustering

11. What is the elbow method in K-Means clustering?

->The Elbow Method helps choose the best number of clusters (K).

Steps:

Run K-Means for different values of K.

Plot K vs Inertia.

Look for the point where inertia reduces slowly (the "elbow").

That K is considered the optimal number of clusters.

12.Describe the concept of "density" in DBSCAN.

->In DBSCAN, density refers to how many data points lie in a small region.

A region is considered dense if:

A point has at least minPts neighbors

Within a radius ε (epsilon)

Dense regions become clusters. Sparse regions become noise/outliers.

13.Can hierarchical clustering be used on categorical data?

->Yes, but not directly with distance measures like Euclidean.

To use hierarchical clustering on categorical data:

Convert to suitable distance metric (e.g., Hamming distance, Jaccard distance), or

Use algorithms designed for categorical data (e.g., k-modes, hierarchical clustering with Gower distance**).

So: possible, but requires special distance measures.

14.What does a negative Silhouette Score indicate?

->A negative Silhouette Score means:

Data points are closer to other clusters than to their own cluster.

The clustering is poor or incorrect.

Clusters overlap badly or K is wrongly chosen.

15.Explain the term "linkage criteria" in hierarchical clustering?

->Linkage criteria define how the distance between clusters is measured when merging them.

Common types:

Single Linkage → min distance between points of two clusters

Complete Linkage → max distance

Average Linkage → average pairwise distance

Ward’s Method → minimizes variance within clusters

Linkage controls the structure and shape of the dendrogram.

16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

->K-Means performs poorly on data where clusters have different sizes, densities, or shapes because:

1. K-Means Assumes Equal-Sized, Spherical Clusters

K-Means uses Euclidean distance and centroid averaging, which naturally forms round and similarly sized clusters.
It cannot correctly separate:

One large cluster and one small cluster

One dense cluster and one sparse cluster

2. Centroid Position Gets Pulled Toward Large or Dense Clusters

Large or dense clusters have more points, so their centroid attracts nearby smaller clusters, causing incorrect assignments.

3. Points From Sparse Clusters Get Assigned to Dense Clusters

Since K-Means optimizes distance to centroids, it often pushes points from a low-density cluster into a high-density cluster.

4. Fixed K Value Cannot Capture Density Variation

Different densities require different cluster definitions, but K-Means uses a single global density assumption.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

->DBSCAN uses two key parameters:

(1) ε (Epsilon) – Neighborhood Radius

Defines how close points must be to be considered neighbors.

Small ε → many small clusters, more noise

Large ε → fewer large clusters, merges everything

(2) minPts – Minimum Points Required to Form a Dense Region

Defines how many points must lie within ε to form a core point.

Low minPts → detects small clusters but may include noise

High minPts → stable clusters but may miss small ones

Influence on Clustering

Points inside a dense ε-neighborhood become core points.

Regions reachable through core points form clusters.

Points that are not core or border → noise/outliers.

18.How does K-Means++ improve upon standard K-Means initialization?

->K-Means++ improves cluster initialization by selecting better initial centroids.

Working:

Choose the first centroid randomly.

For each remaining centroid, select a point that is farther from already chosen centroids (probability ∝ distance²).

Run K-Means normally.

Advantages:

Avoids random initialization trap.

Reduces chances of bad clustering.

Faster convergence.

Produces more stable and accurate clusters.

19.What is agglomerative clustering?

->Agglomerative clustering is a type of hierarchical clustering that works in a bottom-up manner.

Steps:

Start with each data point as an individual cluster.

Merge the two closest clusters based on a distance measure.

Repeat until only one cluster remains (or desired clusters).

Features:

Uses linkage criteria (single, complete, average, Ward’s).

Produces a dendrogram.

Does not require specifying K initially.

20.What makes Silhouette Score a better metric than just inertia for model evaluation?

->Inertia:

Only measures compactness of clusters.

Always decreases as K increases → cannot judge cluster separation.

Does not detect overfitting or overlapping clusters.

Silhouette Score:

Measures both:

Cohesion: How close a point is to its own cluster

Separation: How far it is from other clusters

Range: −1 to +1

Higher score → well-formed, well-separated clusters

Helps choose the best K and evaluate cluster quality.

Why better?

Silhouette Score evaluates overall clustering quality, not just compactness, making it a more reliable metric than inertia alone.

```python
'''29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart'''

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=500, centers=4, random_state=42)

ks = [2, 3, 4, 5]
scores = []

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.bar(ks, scores)
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
'''31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries'''
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.8, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(7, 5))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

```python
'''34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot'''

from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data

ks = [2, 3, 4, 5, 6]
inertias = []

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(ks, inertias, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("K-Means Inertia for K = 2 to 6")
plt.show()
```
